```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 1, 100)

f_sin = np.sin(x)
f_cos = np.cos(x)

fig, axes = plt.subplots(1, 2)
plt.subplots_adjust(wspace=0.5)

axes[0].plot(x, f_sin)
axes[0].set_title("sin(x)")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

axes[1].plot(x, f_cos)
axes[1].set_title("cos(x)")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.savefig("sincospanels.pdf", bbox_inches="tight")
plt.show()
```
